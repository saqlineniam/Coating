```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\saqli\Desktop\DATA for Strawberry\Brix.xlsx')
```

```python
df
```

```python
df = df[['Treatments', 'Storage Time', 'Unnamed: 3','Unnamed: 4']]
```

```python
df = df.rename(columns= {'Treatments':'Treatments', 'Storage Time':'Day 2', 'Unnamed: 3':'Day 4','Unnamed: 4': 'Day 6'})
```

```python
df = df.drop(0)
```

```python
df
```

```python
df
```

```python
df = df.reset_index(drop = True)
```

```python
df_melt = df.melt(id_vars=['Treatments'],
        value_vars=['Day 2', 'Day 4', 'Day 6'],
        var_name = 'Days',
        value_name = "Brix"
        )
```

```python
df_melt
```

```python
df_melt = df_melt.pivot(index = 'Days', columns = 'Treatments', values = 'Brix')
```

```python
ax = df_melt.plot(style=['o-', 's-', '^-', 'D-', 'v-', '<-', '>-', 'p-', 'h-', '*-', 'X-'], markersize=8)
ax.set_ylabel('Firmness (Nm)')
ax.set_title('Firmness Values Over Time by Treatment')
ax.legend(title='Treatments', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.gcf().set_size_inches(10, 6)
```
